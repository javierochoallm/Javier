**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

sp.var("x")

def f(x):
    return 1260 + 117*x - 9*x**2
df = sp.diff(f(x))
df
pc = sp.solve(df, x)
pc
```

```python
dff = sp.diff(df, x)
dff
```

```python
for c in pc:
    crit_2da_derivada = dff.subs(x, c)
    print(crit_2da_derivada)
```

```python
xn = np.linspace (-100, 100, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc : 
    plt.plot(c, f(c), "*r", ms = 10)
    
    
print("El punto critico es un maximo global")
```

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
dff = sp.diff(df, x)
dff
```

```python
for c in pc:
    crit_2da_derivada = dff.subs(x, c)
    print(crit_2da_derivada)
```

```python
xn = np.linspace (-10, 10, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc : 
    plt.plot(c, f(c), "*r", ms = 10)
    
    
print("El punto critico", pc "es un minimo local")
```

$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python

def f(x):
    return 6 - 2*x + (18-2*x**2)**0.5
df = sp.diff(f(x))
df
pc = sp.solve(df, x)
pc
```

```python
xn = np.linspace (0.1, 10, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc : 
    plt.plot(c, f(c), "*r", ms = 10)
    
    
print("El punto critico es un minimo local")
```
